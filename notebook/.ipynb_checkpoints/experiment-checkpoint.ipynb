```python
df = df.drop(columns=["record_id"])
```

```python
X = df.drop(columns=["gaming_addiction_risk_level"])
y = df["gaming_addiction_risk_level"]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42,stratify=y)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
columns = num.columns
```

```python
columns
```

```python
scale = StandardScaler()
```

```python
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
```

```python
X_train_scaled[columns] = scale.fit_transform(X_train[columns])
X_test_scaled[columns] = scale.transform(X_test[columns])
```

```python
X_train_scaled.shape
```

```python
X_train_scaled.head(3)
```

```python
X_test_scaled.head(3)
```

```python
knn = KNNImputer(n_neighbors=5)
```

```python
X_train_imputed = X_train_scaled.copy()
X_test_imputed = X_test_scaled.copy()
```

```python
X_train_imputed[columns] = knn.fit_transform(X_train_scaled[columns])
X_test_imputed[columns]  = knn.transform(X_test_scaled[columns])
```

```python
X_train_imputed.shape
```

```python
X_train_imputed
```

```python
categorical = X_train_imputed.select_dtypes(include=["object"])
cat = categorical.columns
```

```python
cat
```

```python
for col in cat:
    print(f"{col} : {X_train_imputed[col].unique()}") 
```

```python
ordinal_cols = ["sleep_quality","academic_work_performance","sleep_disruption_frequency","mood_swing_frequency"]
```

```python
ordinal_categories = [
    ["Insomnia","Very Poor", "Poor", "Fair", "Good"],
    ["Failing", "Poor", "Below Average", "Average", "Good", "Excellent"],
    ["Never", "Rarely", "Sometimes", "Often", "Always"],
    ["Never", "Rarely", "Sometimes", "Often", "Daily"]
]
```

```python
ohe_cols = cat.difference(ordinal_cols)
```

```python
ohe_cols
```

```python
transform = ColumnTransformer(transformers=[
    ("ohe",OneHotEncoder(drop="first",handle_unknown="ignore",sparse_output=False),ohe_cols),
    ("ordinal",OrdinalEncoder(categories=ordinal_categories),ordinal_cols)
],remainder="passthrough")

transform.set_output(transform="pandas")
```

```python
X_train_transform  = transform.fit_transform(X_train_imputed)
X_test_transform = transform.transform(X_test_imputed)
```

```python
X_train_transform.columns = X_train_transform.columns.str.replace(r'^(ohe|ordinal|remainder)__','',regex=True)
X_test_transform.columns = X_test_transform.columns.str.replace(r'(ohe|ordinal|remainder)__','',regex=True)
```

```python
le = LabelEncoder()
y_train_label = le.fit_transform(y_train)
y_test_label = le.transform(y_test)
```

```python
X_train_transform
```

```python
model = RandomForestClassifier(
    n_estimators = 300,
    min_samples_split = 10,
    min_samples_leaf = 3,
    max_depth = 7
)
```

```python
model.fit(X_train_transform,y_train_label)
```

```python
model.score(X_test_transform,y_test_label)
```

```python
result = permutation_importance(
    model,
    X_test_transform,
    y_test_label,
    n_repeats = 10,
    random_state = 42
)
```

```python
for i in result.importances_mean.argsort()[::-1]:
    print(X_test_transform.columns[i], result.importances_mean[i]*100)
```
